### **Zadanie 5.**

Dla funkcji $f(x)$ zdefiniowanej w sekcji **Cel zajęć** znajdź miejsca zerowe przy użyciu:

- funkcji [`sp.optimize.root()`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.root.html).
- funkcji [`sp.optimize.fsolve()`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.fsolve.html).

```python
# ===== Twoja implementacja tutaj =====
fun = main.func
x1_start_guess = -0.5
x2_start_guess = 1
x1_root = sp.optimize.root(fun, x1_start_guess).x
x2_root = sp.optimize.root(fun, x2_start_guess).x
x1_fsolve = sp.optimize.fsolve(fun, x1_start_guess)
x2_fsolve = sp.optimize.fsolve(fun, x2_start_guess)
print('Miejsca zerowe znalezione przy pomocy funkcji roots: \n{0}, {1}\n'.format(x1_root, x2_root))
print('Miejsca zerowe znalezione przy pomocy funkcji fsolve: \n{0}, {1}\n'.format(x1_fsolve, x2_fsolve))
if x1_fsolve == x1_root:
    print('Uzyskane miejsca zerowe x1 SĄ równe!')
else:
    print('Uzyskane miejsca zerowe x1 nie SĄ równe!')
    print('Różnica wynosi: {0}'.format(abs(x1_root-x1_fsolve)))
if x2_fsolve == x2_root:
    print('Uzyskane miejsca zerowe x2 SĄ równe!')
else:
    print('Uzyskane miejsca zerowe x2 nie SĄ równe!')
    print('Różnica wynosi: {0}'.format(abs(x2_root-x2_fsolve)))
```

### **Zadanie 6.**

Korzystając z przedziałów wyznaczonych w **Zadaniu 1.** znajdź miejsca zerowe funkcji $f(x)$ zdefiniowanej w sekcji **Cel zajęć**, przy użyciu:

- metody bisekcji,
- metody siecznych,
- metody Newtona,

z tolerancją równą $1\mathrm{e}{-10}$.

Zbadaj dokładność (względem rozwiązania z **Zadania 5.**) i czas obliczeń metod w zależności od liczby 100acji. Wyniki przedstaw na wykresach. 

```python
# ===== Twoja implementacja tutaj =====
iteration_values = np.linspace(2,100,20)
time_bisection_x1 = []
time_newton_x1 = []
time_secant_x1 = []
time_bisection_x2 = []
time_newton_x2 = []
time_secant_x2 = []

norms_b = []
norms_s = []
norms_n = []
for iter in iteration_values:
    time = %timeit -r 5 -n 10 -q -o main.bisection(-0.5, 0.3, main.func, 1e-10, 100)
    time_bisection_x1.append(time.average)
    norms_b.append(main.bisection(-0.5, 0.3, main.func, 1e-10, 100))
    time = %timeit -r 5 -n 10 -q -o main.bisection(0.5, 1, main.func, 1e-10, 100)
    time_bisection_x2.append(time.average)

    time = %timeit -r 5 -n 10 -q -o main.secant(-0.5, 0.3, main.func, 1e-10, 100)
    time_secant_x1.append(time.average)
    norms_s.append(main.secant(-0.5, 0.3, main.func, 1e-10, 100))
    time = %timeit -r 5 -n 10 -q -o main.secant(0.5, 1, main.func, 1e-10, 100)
    time_secant_x2.append(time.average)

    time = %timeit -r 5 -n 10 -q -o main.newton(main.func, main.dfunc, main.ddfunc, -0.5, 0.3, 1e-10, 100)
    time_newton_x1.append(time.average)
    norms_n.append(main.newton(main.func, main.dfunc, main.ddfunc, -0.5, 0.3, 1e-10, 100))
    time = %timeit -r 5 -n 10 -q -o main.newton(main.func, main.dfunc, main.ddfunc, 0.5, 1, 1e-10, 100)
    time_newton_x2.append(time.average)


fig, ax = plt.subplots(2, 1, figsize=(10,10))
ax[0].plot(iteration_values, time_bisection_x1, 'o', label='Metoda bisekcji')
ax[0].plot(iteration_values, time_newton_x1, 'o', label='Metoda Newtona')
ax[0].plot(iteration_values, time_secant_x1, 'o', label='Metoda Siecznych')

ax[0].set(title='Dla pierwszego miejsca zerowego',
xlabel='Liczba iteracji',
ylabel='Średni czas potrzebny do otrzymania wyniku [s]')
ax[0].legend()
ax[0].grid()
ax[1].plot(iteration_values, time_bisection_x2, 'o', label='Metoda bisekcji')
ax[1].plot(iteration_values, time_newton_x2, 'o', label='Metoda Newtona')
ax[1].plot(iteration_values, time_secant_x2, 'o', label='Metoda Siecznych')

ax[1].set(title='Dla drugiego miejsca zerowego',
xlabel='Liczba iteracji',
ylabel='Średni czas potrzebny do otrzymania wyniku [s]')
ax[1].legend()
ax[1].grid()
fig.suptitle('Zestawienie złożoności czasowej obu funkcji w zależności od liczby iteracji')
plt.show()
fig, ax = plt.subplots(2, 1, figsize=(10,10))
ax[0].axhline(x1_fsolve, label="Wartość oczekiwana")
print(norms_b)
f = [x[0] for x in norms_b]
ax[0].plot(iteration_values, f, "o",label="Metoda bisekcji")
f = [x[0] for x in norms_n]
ax[0].plot(iteration_values, f, "o", label="Metoda Newtona")
f = [x[0] for x in norms_s]
ax[0].plot(iteration_values, f, "o", label="Metoda Siecznych")
ax[0].legend()
ax[1].axhline(x2_fsolve, label="Wartość oczekiwana")
f = [x[1] for x in norms_b]
ax[1].plot(iteration_values, f, "o",label="Metoda bisekcji")
f = [x[1] for x in norms_n]
ax[1].plot(iteration_values, f, "o", label="Metoda Newtona")
f = [x[1] for x in norms_s]
ax[1].plot(iteration_values, f, "o", label="Metoda Siecznych")
ax[1].legend()
plt.show()
```
